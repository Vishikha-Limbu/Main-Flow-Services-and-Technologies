```python
questions = ["1. How many accidents occured on different days of the week?",
             "2. What is the distribution of accidents  by age band of the driver?",
             "3. What is the distribution of accidents by sex of the driver?",
             "4. What are the major causes of accidents?"
            ]
questions
```

```python
#Let's find the answer of first question.

#1.How many accidents occured on different days of the week?

#getting the value
df['Day_of_week'].value_counts()
```

```python
#plotting bar chart.
df['Day_of_week'].value_counts().plot(kind = 'bar', color=["orchid"])
plt.title("Accidents by Day of the week")
plt.xlabel("Day of the week")
plt.ylabel("Number of Accidents");
```

```python
#plotting a pie chart
df['Day_of_week'].value_counts().plot(kind = 'pie', figsize = (8,6))
plt.title("Accidents by day of the week")
plt.ylabel("")
plt.legend(df["Day_of_week"].unique());
```

```python
#Let's find the answer of our second question.

#2.What is the distribution of accidents  by age band of the driver?'

df['Age_band_of_driver'].value_counts()
```

```python
#plotting the bar chart
df['Age_band_of_driver'].value_counts().plot(kind='bar',color=['salmon','lightskyblue','springgreen','khaki','coral','plum'])
plt.title("Accidents by Age Band of Driver")
plt.xlabel("Age Band of Driver")
plt.ylabel("Number of Accidents");
```

```python
#plotting a pie chart 

df['Age_band_of_driver'].value_counts().plot(kind='pie',figsize=(8,6))
plt.title("Accidents by Age Band of Driver")
plt.ylabel("")
plt.legend(df['Age_band_of_driver'].unique());
```

```python
#Let's move to question 3

#3.What is the distribution of accidents by sex of the driver?

df['Sex_of_driver'].value_counts()
```

```python
#plotting a bar chart
df['Sex_of_driver'].value_counts().plot(kind='bar',color=['salmon','lightskyblue'])
plt.title("Accidents by Sex of Driver")
plt.xlabel("Sex of Driver")
plt.ylabel("Number of Accidents");
```

```python
#plotting a pie chart
df['Sex_of_driver'].value_counts().plot(kind='pie',figsize=(8,6))
plt.title("Accidents by Sex of Driver")
plt.ylabel("")
plt.legend(df['Sex_of_driver'].unique());

```

```python
#Now question 4!

#4.What are the major causes of accidents?
df['Cause_of_accident'].value_counts()
```

```python
#plotting a bar chart
df['Cause_of_accident'].value_counts().plot(kind='bar',color=['coral','lightskyblue','plum','khaki','green','red','purple'])
plt.title("Causes of Accidents")
plt.xlabel("Cause of Accident")
plt.ylabel("Number of Accident");
```

```python
#plotting a pie chart
df['Cause_of_accident'].value_counts().plot(kind='pie',figsize=(8,6))
plt.title("Causes of Accident")
plt.ylabel("")
plt.legend(df['Cause_of_accident'].unique());


```
